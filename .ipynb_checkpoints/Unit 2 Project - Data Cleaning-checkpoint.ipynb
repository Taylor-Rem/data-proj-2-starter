### Step 4: Create Data Dictionaries for Each Dataset

```python
####### TO DO ###########

# Work in the cells below to fill in missing information in the data dictionary. 


# File 1: house_data.csv
# The table contains information on various properties such as the number of bedrooms 
# and bathrooms, the square footage of the living area and lot, the number of floors, 
# and whether it has a waterfront view. Other information includes the condition and grade
# of the property, square footage of above and basement areas, year built and renovated, 
# zip code, latitude and longitude coordinates, and square footage of living area and lot in 2015.

# ID:
# Description: Unique identifier for each property
# Data Type: Integer

# Bedrooms:
# Description: Number of bedrooms in the property
# Data Type: Integer

# Bathrooms:
# Description: Number of bathrooms in the property
# Data Type: Float

# SQFT living:
# Description: Square footage of the living area of the property
# Data Type: Integer

# SQFT lot:
# Description: Square footage of the lot on which the property is built
# Data Type: Float

# Floors:
# Description: Number of floors in the property
# Data Type: Float

# Waterfront?:
# Description: Indicates whether the property has a waterfront view (Yes/No)
# Data Type: String

# View:
# Description: Quality of the view from the property (Very Poor, Poor, OK, Good, Excellent)
# Data Type: String

# House Condition:
# Description: Overall condition of the property (OK Condition, Great Condition, Poor Condition, Very Poor Condition)
# Data Type: String

# Grade:
# Description: Evaluative measure of the property's construction and design (1-13)
# Data Type: Float

# SQFT above:
# Description: Square footage of the area of the property above the basement
# Data Type: Integer

# SQFT basement:
# Description: Square footage of the basement in the property
# Data Type: Integer

# Year Built:
# Description: Year in which the property was built
# Data Type: Float

# Year Renovated:
# Description: Year in which the property was last renovated (NaN if not renovated)
# Data Type: Float

# Zipcode:
# Description: Zip code of the property
# Data Type: Integer

# Lat:
# Description: Latitude coordinate of the property
# Data Type: Float

# Long:
# Description: Longitude coordinate of the property
# Data Type: Float

# SQFT living 15:
# Description: Square footage of living area in 2015
# Data Type: Integer

# SQFT lot 15:
# Description: Square footage of the lot in 2015
# Data Type: Integer


# File 2: The table contains information about real estate properties including 
# their unique identifier, date of sale, and sale price. The table has three columns:
# Id, DATE, and Price. The Id column provides a unique identifier for each property. 
# The DATE column includes the date of sale of the property in the format "YYYYMMDDT000000". 
# The Price column shows the sale price of the property, represented in the format "$XXX,XXX.XX".

# Id:
# Description: Unique identifier for each property
# Data Type: Integer

# DATE:
# Description: Date of sale of the property
# Data Type: String (in the format "YYYYMMDDT000000")

# Price:
# Description: Sale price of the property
# Data Type: String (in the format "$XXX,XXX.XX")
```
